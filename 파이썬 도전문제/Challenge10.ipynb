# 문제 1

길이가 n인 배열이 하나 있으며, 상혁이는 현재 첫 번째 칸에 위치하고 있다.

각 칸에는 1이상 n이하의 숫자가 하나씩 적혀 있으며, 각 칸에서 오른쪽으로 해당 숫자만큼 떨어진 칸으로 움직일 수 있다.

만약 해당 숫자만큼 떨어진 칸이 배열을 벗어나 존재하지 않으면 이동할 수 없다.

상혁이는 첫 번째 칸에서 n번째 칸까지 이동하고자 한다. n번째 칸에 도착할 수 있다면 도착에 필요한 이동 횟수를 반환하고, 도착할 수 없다면 -1를 반환하여라

### 제한

2 ≤ n ≤ 10

1 ≤ 각 칸에 적혀 있는 숫자 ≤ n

n, 각 칸에 적혀 있는 숫자는 모두 자연수

### 입력

배열의 길이 n와 배열 정보 arr가 주어진다.

### 예제 입력

n = 7

arr = [2, 2, 1, 3, 3, 4, 2] 

### 예제 출력

3

```python
# Your code here
# Your code here
def move_list(n,arr):
    idx=0
    count=0
    if n<2 or n>10:
        raise print('n은 2<=n<=10')
    for i in arr:
        if i is not int or i<1:
            raise print('배열안에 숫자는 모두 자연수여야 합니다.')
    while True:
        try:
            idx+=arr[idx]
            count+=1
            if idx==len(arr)-1:
                print(count)
                break
        except IndexError:
            print(-1)
            break

n=7
arr=[2,2,1,3,3,4,2]
move_list(n,arr)
```

# 문제 2 (Hard Ver.)

길이가 n인 배열이 하나 있으며, 한솔이는 현재 첫 번째 칸에 위치하고 있다.

각 칸에는 1이상 n이하의 숫자가 하나씩 적혀 있으며, 각 칸에서 오른쪽으로 해당 숫자만큼 떨어진 칸으로 움직일 수 있다. **단, 딱 한 번 왼쪽으로 해당 숫자만큼 떨어진 칸으로 이동할 수 있으며, 왼쪽으로 이동할 수 있는 기회를 사용하지 않아도 된다.**

만약 해당 숫자만큼 떨어진 칸이 배열을 벗어나 존재하지 않으면 이동할 수 없다.

한솔이는 첫 번째 칸에서 n번째 칸까지 총 3번 이내로 움직여서 도착하고자 한다. 이러한 방법이 존재하면 ‘YES’를, 그렇지 않으면 ‘NO’를 출력하여라.

### 제한

2 ≤ n ≤ 10

1 ≤ 각 칸에 적혀 있는 숫자 ≤ n

n, 각 칸에 적혀 있는 숫자는 모두 자연수

### 입력

배열의 길이 n와 배열 정보 arr가 주어진다.

### 출력

첫 번째 칸에서 n번째 칸까지 총 3번 이내로 움직여서 도착할 수 있는 방법이 존재하면 ‘YES’를, 그렇지 않으면 ‘NO’를 출력하여라.

### 예제 입력

n = 7

arr = [2, 2, 1, 3, 3, 4, 2] 

### 예제 출력

YES

```python
# your code here
def move_list2(n,arr):
    idx=0
    count=0
    plus_one=0
    if n<2 or n>10:
        raise 'n은 2<=n<=10'
    elif n!=len(arr):
        raise 'n은 배열의 길이와 동일해야 합니다.'
    for i in arr:
        if i!=int(i) or i<1:
            raise '배열안에 숫자는 모두 자연수여야 합니다.'
    while True:
        if count>3:
            print('NO')
            break
        try:
            idx+=arr[idx]
            count+=1
            if idx==len(arr)-1:
                print('Yes')
                break
            elif idx-arr[idx]+arr[idx-arr[idx]]==len(arr)-1 and count<=1 and plus_one==0:
                print('Yes')
                break
```
